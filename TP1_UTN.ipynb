```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta, date
```

```python
# 1. Crear una lista con todas las fechas utilizando la libreria Datetime
dia_inicio = date(2022, 10, 1)
dia_fin = date(2023, 10, 1)
delta = timedelta(days=1)
dias = []
while dia_inicio <= dia_fin:
    dias.append(dia_inicio)
    dia_inicio += delta
print(dias)
```

```python
# 2. Crear una lista con temperaturas aleatorias en grados Fahrenheit
np.random.seed(0)  # 4. Setear una semilla
temps_fahrenheit = np.random.uniform(low=14, high=104, size=len(dias))#Temperaturas creadas aleatoriamnte (valores flotantes)
temps_fahrenheit = [int(valor) for valor in temps_fahrenheit] #convierto los valores flotantes creados aleatoriamente a enteros
print(temps_fahrenheit)
```

```python
# 3. Crear una lista con valores de presiÃ³n atmosfÃ©rica en pascales
presiones_pascal = np.random.uniform(low=10000, high=105000, size=len(dias)) #rango de valores en Pascales para la zona da Argentina
presiones_pascal = [int(valor2) for valor2 in presiones_pascal] #valores flotantes creados aleatoriamente a enteros
print(presiones_pascal)
```

```python
# 5. Crear una funciÃ³n para pasar la lista de temperaturas a grados Celsius y la de presiones a atmÃ³sferas
def fahrenheit_a_celsius(temperaturas_fahrenheit):
        return [round(((temp - 32) / 1.8), 1) for temp in temperaturas_fahrenheit]

def pascal_a_atmospheres(presion_en_pascales):
    return [round((atm / 101325),1) for atm in presion_en_pascales] # 101325 valor aprox a 1 atmosfera: rango normal para Argentina de 01 a 1 atm

temps_celsius = fahrenheit_a_celsius(temps_fahrenheit)
presiones_atm = pascal_a_atmospheres(presiones_pascal)

print(temps_celsius)
print()
print(presiones_atm)
```

```python
# 6. Unificar las listas creadas en un DataFrame
df_temp_y_pres = pd.DataFrame({
    'Fecha': dias,
    'Temperaturas': temps_celsius,
    'Presion_atms': presiones_atm
})

df_temp_y_pres
```

```python
# 7. Hacer un anÃ¡lisis de integridad
print(df_temp_y_pres.info()) # Tipos de datos que contiene el df
print()
print ("*"*50)
print(df_temp_y_pres.describe()) # Estadistica del df
print()
print ("*"*50)
#-----------------------------------------------------------------------
#El dia con la temperatura mas alta
print(df_temp_y_pres.groupby("Fecha")["Temperaturas"].sum().idxmax()) # con idmax() se encuentra el indice del valor maximo de la columna "Temperaturas" para la consulta en "Fecha"
print()

#Confirmacion 1 dia con la temperatura maxima
print(df_temp_y_pres.loc[df_temp_y_pres['Temperaturas'].idxmax()])
print()

#Confirmacion 2 dia con la temperatura maxima
dia_max_temperatura = df_temp_y_pres.loc[df_temp_y_pres['Temperaturas'].idxmax(), 'Fecha']
print(f"El dÃ­a con la temperatura mÃ¡xima es: {dia_max_temperatura}")
print()
print ("*"*50)
#-----------------------------------------------------------------------
#El dia con la temperatura mas baja
print(df_temp_y_pres.groupby("Fecha")["Temperaturas"].sum().idxmin()) # con idmin() se encuentra el indice del valor minimo de la columna "Temperaturas" para la consulta en "Fecha"
print()

#Confirmacion 1 dia con la temperatura minima
print(df_temp_y_pres.loc[df_temp_y_pres['Temperaturas'].idxmin()])
print()

#Confirmacion 2 dia con la temperatura minima
dia_temperatura_min = df_temp_y_pres.loc[df_temp_y_pres['Temperaturas'].idxmin(), 'Fecha']
print(f"El dÃ­a con la temperatura minima es: {dia_temperatura_min}")
print()
print ("*"*50)
#-----------------------------------------------------------------------
```

```python
# 8. Realizar los siguientes grÃ¡ficos
# a. GrÃ¡ficos de lÃ­neas para ver las temperaturas y presiones a lo largo del tiempo
df_temp_y_pres.set_index("Fecha")[["Temperaturas", "Presion_atms"]].plot(subplots=True)
plt.show()
```

```python
# b. Un diagrama de dispersiÃ³n entre temperatura y presiÃ³n
plt.scatter(df_temp_y_pres["Temperaturas"], df_temp_y_pres["Presion_atms"])
plt.xlabel("Temperatura (Celsius)")
plt.ylabel("Presion (Atmospheres)")
plt.show()
```

```python
# c. Un boxplot para cada variable continua para detectar outliers
df_temp_y_pres[["Temperaturas", "Presion_atms"]].plot(kind='box', subplots=True)
plt.show()
```

```python
# d. Matriz de correlaciÃ³n
corr = df_temp_y_pres[["Temperaturas", "Presion_atms"]].corr()
sns.heatmap(corr, annot=True)
plt.show()
```
